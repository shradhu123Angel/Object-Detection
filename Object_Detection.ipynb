```python
import cv2

img = cv2.imread("/content/working office.jpg")
```

```python
from google.colab.patches import cv2_imshow

cv2_imshow(img)
```

```python
!git clone https://github.com/pysource7/simpledetection
```

```python
# Object detection
from simpledetection.objectdetection import ObjectDetection

od = ObjectDetection()
```

```python
class_ids, scores, bboxes = od.detect(img)

# count specific object
person = 0
laptop = 0
book = 0
for class_id, score, box in zip(class_ids, scores, bboxes):

    (x, y, w, h) = box
    print(class_id)
    class_name = od.classes[class_id[0]]

    if class_name == "person":
        person += 1
    elif class_name == "laptop":
        laptop += 1
    elif class_name == "book":
        book += 1
    color = od.colors[class_id[0]]
    cv2.putText(img, class_name, (x, y - 5), 0, 1, color, 2) # -5 name above the image
    cv2.rectangle(img, (x,y), (x + w, y + h), color, 3)

cv2_imshow(img)
```

```python
print(person)
print(book)
print(laptop)

with open("summary.txt", "w") as fo:
    fo.write("In this picture there are\n")
    fo.write(str(laptop) + " laptops")
    
    fo.write(str(book) + " books")
    fo.write(str(person) + " people")
```
